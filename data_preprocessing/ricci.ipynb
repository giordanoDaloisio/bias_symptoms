```python
import pandas as pd
```

```python
data = pd.read_csv('raw_data/ricci.csv', index_col=0)
```

```python
data.drop(columns='X', inplace=True)
```

```python
data = pd.get_dummies(data)
```

```python
data
```

```python
data['Combine'] = data['Combine'].apply(lambda x: x >= 60)
```

```python
data['Combine'] = data['Combine'].astype(int)
```

```python
data.to_csv('new_data/ricci.csv')
```
