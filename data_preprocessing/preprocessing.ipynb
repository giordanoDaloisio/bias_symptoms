```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder

base_path = "raw_data"
end_path = "new_data"
encoder = LabelEncoder()
```

Preprocessed data:

- Ad dataset: label y pos label 1
- Athlete: label Medal_Gold pos label 1
- Diabetic data: label diabetesMed pos label 1
- ibm: label Attrition pos label 0
- nursery: label y pos label 2
- placement: label status pos label 1
- vaccine: label lowtrustvaccinerec label 0
- us: label dIncome1 3

```python
## Ad dataset

ad = pd.read_csv(f'{base_path}/ad.data', header=None, low_memory=False, na_values='   ?')
ad.rename(columns={1558: 'y'}, inplace=True)
ad = ad.dropna()
ad.replace('?', np.nan, inplace=True)
ad = ad.dropna()
ad['y'] = encoder.fit_transform(ad['y'])
ad.to_csv(f'{end_path}/ad.csv')
```

```python
# Athlete

athlete = pd.read_csv(f'{base_path}/athlete_events.csv', index_col=0)
athlete.dropna(inplace=True)
athlete.drop(columns=['Name', 'Team', 'Games', 'Event'], inplace=True)
athlete = pd.get_dummies(athlete, dtype=np.int64)
athlete.to_csv(f'{end_path}/athlete.csv')
```

```python
# Diabetes

diabetes = pd.read_csv(f'{base_path}/diabetes.csv')
diabetes.head()
```

```python
# Diabetic data

diabetic = pd.read_csv(f'{base_path}/diabetic_data.csv', na_values='?', index_col=0)
diabetic.drop(columns=['patient_nbr','weight', 'payer_code', 'medical_specialty'], inplace=True)
diabetic.dropna(inplace=True, axis=1)
diabetic['diabetesMed'] = encoder.fit_transform(diabetic['diabetesMed'])
diabetic = pd.get_dummies(diabetic, dtype=np.int64)
diabetic.to_csv(f'{end_path}/diabetic.csv')
```

```python
# ibm

ibm = pd.read_csv(f'{base_path}/ibm.csv')
ibm['Attrition'] = encoder.fit_transform(ibm['Attrition'])
ibm = pd.get_dummies(ibm, dtype=np.int64)
ibm.to_csv(f'{end_path}/ibm.csv')
```

```python
# nursery

nursery = pd.read_csv(f'{base_path}/nursery.data', header=None)
nursery.rename(columns={8: 'y'}, inplace=True)
nursery['y'] = encoder.fit_transform(nursery['y'])
nursery = pd.get_dummies(nursery, dtype=np.int64)
nursery.to_csv(f'{end_path}/nursery.csv')
```

```python
# placement

placement = pd.read_csv(f'{base_path}/Placement_Data_Full_Class.csv', index_col=0)
placement['status'] = encoder.fit_transform(placement['status'])
placement.drop(columns=['salary'], inplace=True)
placement = pd.get_dummies(placement, dtype=np.int64)
placement.to_csv(f'{end_path}/placement.csv')
```

```python
# vaccine

vaccine = pd.read_excel(f'{base_path}/vaccine.xlxs')
vaccine.dropna(inplace=True)
vaccine.drop(columns=['SND_study', 'SND_dataset', 'SND_version'], inplace=True)
vaccine.set_index('id', inplace=True)
vaccine.to_csv(f'{end_path}/vaccine.csv')
```

```python
# us census

us = pd.read_csv(f'{base_path}/USCensus1990.data.txt', index_col=0)
us.dropna(inplace=True)
us.drop(columns=['dIncome2', 'dIncome3'], inplace=True)
us.to_csv(f'{end_path}/us.csv')
```
