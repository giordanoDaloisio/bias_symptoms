```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")
```

```python
times = pd.read_fwf('training_times.txt', header=None)
```

```python
fig = sns.boxplot(times[152:])
plt.title('Training times for the COMPAS dataset')
plt.xticks('')
plt.xlabel('Logistic Regression')
plt.ylabel('Training time (s)')
plt.savefig('training_times.png', dpi=300)
plt.savefig('training_times.pdf')
plt.show()
```
