```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

sns.set_theme(style="whitegrid", font_scale=1.5)
```

```python
symptoms = pd.DataFrame()
for file in os.listdir("symptoms"):
  data = pd.read_csv("symptoms/" + file, index_col=0)
  symptoms = pd.concat([symptoms, data])
  
symptoms = symptoms[symptoms['importance']>0.05]
```

```python
corr = symptoms.corr()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr, annot=True, cmap="coolwarm")
plt.show()
```
